```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier,plot_tree,export_graphviz
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
df=pd.read_csv('titanic_train.csv')
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df=df.drop(['PassengerId','Name','Ticket','Fare','Cabin'],axis=1)
df
```

```python
embarked=pd.get_dummies(df['Embarked'],drop_first=True)
embarked
```

```python
sex=pd.get_dummies(df['Sex'],drop_first=True)
sex
```

```python
df=pd.concat([df,sex,embarked],axis=1)
df
```

```python
df=df.drop(['Sex','Embarked'],axis=1)
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.fillna(df['Age'].median().astype('int64'))
df
```

```python
df.isnull().sum()
```

```python
x=df.drop('Survived',axis=1)
y=df['Survived']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
hyperparameters={'criterion':['gini','entropy'],'max_depth':list(range(2,10))}
```

```python
dtc=DecisionTreeClassifier()
```

```python
best_model=GridSearchCV(dtc,hyperparameters,cv=5)
best_model.fit(x_train,y_train)
y_pred_best=best_model.predict(x_test)
best_model.best_params_
```

```python
best_model.score(x_test,y_test)
```

```python
dtc.fit(x_train,y_train)
```

```python
y_pred=dtc.predict(x_test)
y_pred
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
plt.figure(figsize=(15,20))
plot_tree(decision_tree=dtc,rounded=True,filled=True,class_names=x.columns)
export_graphviz(dtc)
```
